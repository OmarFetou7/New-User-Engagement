```python
users['crt_seq_day'] = (users["Created At Month"]-1) * 31 + users["Created At Day_of_month"]
activity['act_seq_day'] = (activity["datetime Month"]-1) * 31 + activity["datetime Day_of_month"]
tmp = users.merge(activity[['User_ID', 'Title', 'act_seq_day']], on= 'User_ID', how='left')
tmp.shape
tmp['target'] = tmp.act_seq_day.between(tmp.crt_seq_day +31 , tmp.crt_seq_day + 62, inclusive= 'neither')
tmp['training_window'] = tmp["act_seq_day"].between(tmp['crt_seq_day'], tmp['crt_seq_day'] + 31, inclusive= 'both')
tmp.drop(tmp.index[tmp.crt_seq_day > 341] ,axis = 0, inplace=True)
print("count active in 2nd month:", tmp[tmp['target'] == True]["User_ID"].unique().shape)
tmp.head()
```

```python
act_agg = tmp.groupby("User_ID").agg(
        target=("target", "max"),          # if any True â†’ 1
        n_acts=("training_window", "sum"),   # number of activities
        crt_day = ("crt_seq_day", "first"),
        last_act=("act_seq_day", "max"),   # latest activity day
    )
act_agg["last_act"].fillna(0, inplace=True)
act_agg.shape
```

```python
print(users.shape)
agg1_df = users.merge(act_agg, on = 'User_ID' , how= 'left' )
print(agg1_df.shape)
```
